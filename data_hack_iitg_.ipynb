```python
import pandas as pd
df_x = pd.read_csv("C:\\Users\\Admin\\Downloads\\dataset and all\\training_set_features.csv")
df_y = pd.read_csv("C:\\Users\\Admin\\Downloads\\dataset and all\\training_set_labels.csv")
```

```python
df_x.head()
```

```python
dff = df_x.merge(df_y, on='respondent_id', how='left')
(dff.head())
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

cols_to_scale = [
    'opinion_xyz_vacc_effective',
    'opinion_xyz_risk',
    'opinion_xyz_sick_from_vacc',
    'opinion_seas_vacc_effective',
    'opinion_seas_risk',
    'opinion_seas_sick_from_vacc'
]

scaler = MinMaxScaler(feature_range=(0, 1))
dff[cols_to_scale] = scaler.fit_transform(dff[cols_to_scale])

print(dff[cols_to_scale].head())
```

```python
df_x['age_group'].unique()
```

```python
age_group_means = dff.groupby('age_group')[['xyz_vaccine', 'seasonal_vaccine']].mean()
age_group_means
```

```python
age_group_means['mean_vaccine'] = age_group_means.mean(axis=1)
age_group_mean_mapping = age_group_means['mean_vaccine'].to_dict()
(age_group_mean_mapping)
```

```python
dff['age_group'] = dff['age_group'].map(age_group_mean_mapping)
print(dff['age_group'])
```

```python
dff['education'].unique()
```

```python
dff['education'] = dff['education'].fillna('Unknown')
```

```python
dff['education'].unique()
```

```python
education_mapping = {
    'Unknown': 0,
    '< 12 Years': 0.20,
    '12 Years': 0.40,
    'Some College': 0.60,
    'College Graduate': 0.80
}
dff['education'] = dff['education'].map(education_mapping)
```

```python
import pandas as pd
columns_to_encode = [
    'race', 'sex', 'income_poverty', 'marital_status', 'rent_or_own',
    'employment_status', 'hhs_geo_region', 'census_msa',
    'household_adults', 'household_children', 'employment_industry',
    'employment_occupation'
]
dff = pd.get_dummies(dff, columns=columns_to_encode, drop_first=True)
```
